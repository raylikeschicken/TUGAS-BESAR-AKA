```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable

films = {
    "Action": [
        ("Avatar", 2973706026),
        ("Avengers: Endgame", 2748242781),
        ("Avatar: The Way of Water", 2317514386),
        ("Avengers: Infinity War", 2048359754),
        ("Spider-man: No Way Home", 1921206586),
        ("Jurassic World",1671063641),
        ("The Avengers",1515100211),
        ("Furious 7",1511986364),
        ("Top Gun: Maverick",1461725209),
        ("Avengers: Age of Ultron",1395316979)
    ],
    "Comedy": [
        ("Barbie", 1444999654),
        ("The Hangover Part II", 586764305),
        ("Ted", 556016627),
        ("Kung Fu Panda 4", 547699816),
        ("Meet the Fockers", 516567575),
        ("Les Intouchables", 484630908),
        ("Bruce Almighty", 484468608),
        ("YOLO ()", 479597304),
        ("Home Alone", 476678838),
        ("Successor ()", 469612890)
    ],
    "Horror": [
        ("It", 701012746),
        ("Jaws", 483655893),
        ("It: Chapter Two", 467561586),
        ("The Exorcist", 428824083),
        ("The Nun", 362900037),
        ("Alien: Romulus", 350865342),
        ("Hannibal", 350100280),
        ("A Quiet Place", 334876670),
        ("The Conjuring", 316479150),
        ("Resident Evil: The Final Chapter", 314101190)
    ]
}

# Sorting function
def sort_films_by_revenue(film_list, ascending=True):
    return sorted(film_list, key=lambda x: x[1], reverse=not ascending)

# Rekursif
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

# Iteratif
def fibonacci_iterative(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# Tabel untuk menyimpan data waktu eksekusi
film_table = PrettyTable()
film_table.field_names = ["Film", "Genre", "Box Office", "Recursive Time (s)", "Iterative Time (s)"]


# Program utama
genre = input("Pilih genre (Action, Comedy, Horror): ").capitalize()
if genre not in films:
    print("Genre tidak ditemukan. Program dihentikan.")
    exit()  # Mengakhiri program jika genre tidak valid

# Pilihan sorting
order = input("Sort secara ascending atau descending (asc/desc)? ").lower()
ascending = True if order == 'asc' else False

# Sorting film
sorted_films = sort_films_by_revenue(films[genre], ascending)

recursive_times = []
iterative_times = []

print("\nFilm yang telah diurutkan berdasarkan pendapatan box office:")
for film, revenue in sorted_films:
    start_time = time.time()
    fibonacci_recursive(30)  # Simulasi waktu rekursif
    recursive_time = time.time() - start_time
    recursive_times.append(recursive_time)

    start_time = time.time()
    fibonacci_iterative(30)  # Simulasi waktu iteratif
    iterative_time = time.time() - start_time
    iterative_times.append(iterative_time)

    film_table.add_row([film, genre, f"${revenue:,}", f"{recursive_time:.6f}", f"{iterative_time:.6f}"])

print(film_table)

# Diagram garis untuk waktu eksekusi
plt.figure(figsize=(10, 6))
plt.plot([film[0] for film in sorted_films], recursive_times, label="Recursive Time", marker='o', color='red')
plt.plot([film[0] for film in sorted_films], iterative_times, label="Iterative Time", marker='o', color='green')
plt.title(f"Execution Time Comparison for {genre} Films")
plt.xlabel("Film")
plt.ylabel("Execution Time (s)")
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
